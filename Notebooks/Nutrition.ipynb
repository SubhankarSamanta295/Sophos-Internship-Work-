```python
def data_split(df):
    df_new = pd.DataFrame()

    for feature in df.columns:
        new = df[feature].str.split(" ", n = 1, expand = True)
        new = new.iloc[:,0:1]
        new.columns = [feature]
        df_new = pd.concat([df_new,new],axis=1)
    return df_new
```

```python
data_nutrition_food_new = data_split(df)
```

```python
a = data_nutrition_food[['Food','Serving Size','calories']]
data_nutrition_food_new = pd.concat([a,data_nutrition_food_new],axis=1)
data_nutrition_food_new.head()
```

```python
def enter_your_details_and_food_intake():
    items = int(input('How many food items you consume ? : '))
    
    gender = input("Gender {(Male/male) or (Female/female)}: ")
    if((gender=='Male') or (gender == 'male')):
        gender=0
    elif ((gender=='Female') or (gender == 'female')):
        gender=1
        
    body_weight = float(input('What is your body weight ? : '))
    foods = []
    serving_size = []
    for i in range(0,items):
        food_name = (input('Enter the food name : '))
        food_serving_size = float(input('Enter amount of food intake (gm): '))
        serving_size.append(food_serving_size)
        foods.append(food_name)
    return foods, serving_size, items, body_weight, gender
```

```python
def data_type_convsersion_to_float(df):
    df_new = pd.DataFrame()

    for feature in df.columns:
        new = df[feature].astype(float)
        df_new = pd.concat([df_new,new],axis=1)
    
    return df_new
```

```python
def sum_of_nutrition_consumed(df):
    df_new = pd.DataFrame()
    
    for feature in df.columns:
        new[feature] = df[feature].sum()
        df_new = pd.concat([df_new,new[feature]],axis=1)
    
    return df_new
```

```python
data_nutrition_consumed_final = sum_of_nutrition_consumed(data_food_consumed_final)
data_nutrition_consumed_final
```
